```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib as mpl
from tqdm import trange
import numpy as np
import time
```

```python
content_path = 'res/content_denji.jpg'
style_path = 'res/style3.jpg'

img_path = content_path
```

```python
img_max_size = 1200   # 

img = tf.io.read_file(img_path)

img = tf.image.decode_image(img, channels=3)  # 
img = tf.image.convert_image_dtype(img, tf.float32)  # float32

print(img.shape)
plt.imshow(img)
plt.show()
```

```python
img_shape_ori = tf.cast(tf.shape(img)[:-1], tf.float32)  # float32
print(img_shape_ori)

img_big_size = max(img_shape_ori)  # 

scale = img_max_size / img_big_size  # 
img_shape_new = tf.cast(img_shape_ori * scale, tf.int32)  # 

print('before resize:', img.shape)
img = tf.image.resize(img, img_shape_new)
print('after resize:', img.shape)

img = img[tf.newaxis, :]
print('img shape:', img.shape)
```
